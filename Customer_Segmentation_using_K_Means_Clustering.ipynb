##Optimizing k-mean


```python
wcss = []
cluster_range = range(1,15)
for i in cluster_range:
  km2 = KMeans(n_clusters=i)
  km2.fit(df[['score','income']])
  wcss.append(km2.inertia_)
```

```python
wcss
```

```python
plt.plot(cluster_range, wcss,marker = 'o')
plt.xlabel('cluster')
plt.ylabel('wcss')
```

##After Optimizing

```python
km3 = KMeans(n_clusters=5)
km3.fit(df[['income','score']])
```

```python
df['optimised_clusters']=km3.predict(df[['income','score']])
df.head()
```

```python
center2 = km3.cluster_centers_
center2
```

```python
df0 = df[df.optimised_clusters==0]
df1 = df[df.optimised_clusters==1]
df2 = df[df.optimised_clusters==2]
df3 = df[df.optimised_clusters==3]
df4 = df[df.optimised_clusters==4]

plt.figure(figsize=(10,6))

# Cluster 0
plt.scatter(df0.score, df0.income, color='green', label='cluster 0')

# Cluster 1
plt.scatter(df1.score, df1.income, color='blue', label='cluster 1')

# Cluster 2
plt.scatter(df2.score, df2.income, color='yellow', label='cluster 2')

# Cluster 3
plt.scatter(df3.score, df3.income, color='purple', label='cluster 3')

# Cluster 4
plt.scatter(df4.score, df4.income, color='orange', label='cluster 4')



plt.scatter(center2[:,0], center2[:,1], color='red', marker='*')


plt.xlabel('score')
plt.ylabel('income')

plt.legend()
plt.grid(True)
plt.show()
```
